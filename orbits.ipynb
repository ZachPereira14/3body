```python
import numpy as np
import matplotlib.pyplot as plt

# Gravitational constant multiplied by the mass of the Sun (GM)
GM = 4 * np.pi * np.pi  # AU^3 / day^2

# Planetary masses in Solar masses
planet_masses = {
    "Sun": 1.0,        # Sun
    "Mercury": 1.651e-7,
    "Venus": 2.447e-6,
    "Earth": 3.003e-6,
    "Mars": 3.227e-7,
    "Jupiter": 9.547e-4,
    "Saturn": 2.857e-4,
    "Uranus": 4.365e-5,
    "Neptune": 5.150e-5,
}

# Define the initial conditions for all bodies
bodies = [
    {"name": "Sun", "position": [0, 0], "velocity": [0, 0], "mass": planet_masses["Sun"]},
    {"name": "Mercury", "position": [0.39, 0], "velocity": [0, 10.89], "mass": planet_masses["Mercury"]},
    {"name": "Venus", "position": [0.72, 0], "velocity": [0, 7.35], "mass": planet_masses["Venus"]},
    {"name": "Earth", "position": [1.017, 0], "velocity": [0, 6.179], "mass": planet_masses["Earth"]},
    {"name": "Mars", "position": [1.52, 0], "velocity": [0, 5.03], "mass": planet_masses["Mars"]},
    {"name": "Jupiter", "position": [5.2, 0], "velocity": [0, 2.75], "mass": planet_masses["Jupiter"]},
    {"name": "Saturn", "position": [9.58, 0], "velocity": [0, 2.03], "mass": planet_masses["Saturn"]},
    {"name": "Uranus", "position": [19.22, 0], "velocity": [0, 1.43], "mass": planet_masses["Uranus"]},
    {"name": "Neptune", "position": [30.05, 0], "velocity": [0, 1.14], "mass": planet_masses["Neptune"]},
]

# Mutual gravitational interaction function
def mutual_interaction(positions, masses, id):
    """
    Compute the total acceleration on an object due to all other objects.
    """
    net_force = np.zeros(2)  # Force starts at zero
    for i, (r_other, mass_other) in enumerate(zip(positions, masses)):
        if i == id:
            continue  # Skip self-interaction
        r_diff = r_other - positions[id]  # Vector difference
        distance = np.linalg.norm(r_diff)  # Distance magnitude
        net_force += GM * mass_other * r_diff / (distance ** 3)  # Add gravitational force
    return net_force

# Leapfrog method for integration
def leapfrog(lfdiffeq, r0, v0, t, h):
    """ Leapfrog method using numpy arrays. """
    hh = h / 2.0
    r1 = r0 + hh * lfdiffeq(r0, v0, t)  # 1st: r at h/2 using v0
    v1 = v0 + h * lfdiffeq(r1, v0, t + hh)  # 2nd: v1 using r at h/2
    r1 = r1 + hh * lfdiffeq(r1, v1, t + h)  # 3rd: r1 at h using v1
    return r1, v1

def plot_orbit_all(bodies, time_steps=10000, h=0.001):
    """
    Simulate and plot the orbits of all bodies with mutual interactions.
    """
    num_bodies = len(bodies)
    
    # Initialize positions and velocities
    positions = np.array([body["position"] for body in bodies])
    velocities = np.array([body["velocity"] for body in bodies])
    masses = np.array([body["mass"] for body in bodies])
    
    # Store trajectories for plotting
    trajectories = [np.zeros((time_steps, 2)) for _ in bodies]
    
    # Time loop
    for t in range(time_steps):
        accelerations = np.zeros_like(positions)  # Reset accelerations
        for i in range(num_bodies):
            accelerations[i] = mutual_interaction(positions, masses, i)  # Compute accelerations
        
        # Define the lfdiffeq function (returns acceleration)
        def lfdiffeq(r, v, t):
            return mutual_interaction(r, masses, i)  # For each body i, return the corresponding acceleration
        
        # Apply the leapfrog method (position and velocity update)
        positions, velocities = leapfrog(lfdiffeq, positions, velocities, t * h, h)
        
        # Record positions for plotting
        for i in range(num_bodies):
            trajectories[i][t] = positions[i]
        
        # Debug: print first few positions and velocities
        if t < 10:  # Only print first few time steps
            print(f"Step {t}:")
            for i, body in enumerate(bodies):
                print(f"{body['name']} - Position: {positions[i]}, Velocity: {velocities[i]}")

    # Plot the trajectories
    plt.figure(figsize=(8, 8))
    for i, body in enumerate(bodies):
        trajectory = trajectories[i]
        plt.plot(trajectory[:, 0], trajectory[:, 1], label=body["name"])
    
    # Plot Sun at the origin
    plt.scatter(0, 0, color='yellow', label="Sun", s=100)
    
    # Customize the plot
    plt.title("Orbits with Mutual Interactions")
    plt.xlabel("X Position (AU)")
    plt.ylabel("Y Position (AU)")
    plt.gca().set_aspect('equal', adjustable='box')  # Equal scaling of axes
    plt.legend()
    plt.grid(True)
    plt.show()

# Simulate and plot the orbits
plot_orbit_all(bodies)
```
